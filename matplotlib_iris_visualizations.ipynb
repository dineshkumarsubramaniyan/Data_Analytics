# üìä Matplotlib Visualization with Iris Dataset
This notebook demonstrates various Matplotlib visualizations using the Iris dataset.

```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris

# Load the dataset
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)
df.head()
```

## 1Ô∏è‚É£ Line Plot

```python
plt.plot(df['sepal length (cm)'], label='Sepal Length')
plt.plot(df['sepal width (cm)'], label='Sepal Width')
plt.title("Line Plot of Sepal Dimensions")
plt.xlabel("Index")
plt.ylabel("cm")
plt.legend()
plt.show()
```

## 2Ô∏è‚É£ Bar Plot

```python
mean_values = df.groupby('species').mean()
mean_values[['petal length (cm)', 'petal width (cm)']].plot(kind='bar')
plt.title("Mean Petal Dimensions by Species")
plt.ylabel("cm")
plt.show()
```

## 3Ô∏è‚É£ Scatter Plot

```python
colors = {'setosa': 'r', 'versicolor': 'g', 'virginica': 'b'}
for species in df['species'].unique():
    subset = df[df['species'] == species]
    plt.scatter(subset['sepal length (cm)'], subset['sepal width (cm)'], label=species, color=colors[species])
plt.title("Scatter Plot of Sepal Dimensions")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.legend()
plt.show()
```

## 4Ô∏è‚É£ Histogram

```python
plt.hist(df['petal length (cm)'], bins=20, color='orange', edgecolor='black')
plt.title("Histogram of Petal Length")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Frequency")
plt.show()
```

## 5Ô∏è‚É£ Box Plot

```python
df.boxplot(column=['sepal length (cm)', 'sepal width (cm)'])
plt.title("Boxplot of Sepal Dimensions")
plt.ylabel("cm")
plt.show()
```

## 6Ô∏è‚É£ Pie Chart

```python
species_counts = df['species'].value_counts()
plt.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%', startangle=140)
plt.title("Species Distribution")
plt.show()
```

## 7Ô∏è‚É£ Subplots

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 4))
axs[0].hist(df['sepal length (cm)'], color='skyblue')
axs[0].set_title('Sepal Length')
axs[1].hist(df['sepal width (cm)'], color='salmon')
axs[1].set_title('Sepal Width')
plt.suptitle("Histograms of Sepal Dimensions")
plt.tight_layout()
plt.show()
```
